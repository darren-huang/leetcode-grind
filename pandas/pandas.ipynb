```python
import pandas as pd
import random
```

# Pandas Overview

## data loading

```python
df = pd.DataFrame(
    {
        "group": [1,1,1,2,3,3],
        "a": [1,2,3,4,5,6],
        "b": reversed([1,2,3,4,5,6]),
    }
)
df2 = pd.DataFrame(
    {
        "group": [1,1,1,2,3,3],
        "c": [x*10 for x in [1,2,3,4,5,6]],
        "d": [x*10 for x in reversed([1,2,3,4,5,6])],
        "r": [random.randint(-100, 100) for i in range(6)]
    }
)
```

```python
df1_grouped = df.groupby("group").sum()
df2_grouped = df2.groupby("group").sum()
```

```python
df1_grouped.join(df2_grouped, "group")
```

```python
df_merged = df.merge(df2, on="group", how="outer")
df_merged
```

```python
cols = ["a", "b", "c", "d", "r"]
df_merged["abcdr"] = df_merged[cols].sum(axis=1)
df_sums = df_merged[["group", "abcdr"]]
df_sums
```

```python
group_sums = df_sums.groupby("group").sum().rename(columns={"abcdr": "abcdr_sum"})
merged = df_sums.merge(group_sums, on="group", how="left")
merged["squares"] = (merged["abcdr_sum"] - merged["abcdr"]) ** 2
sqaures_sum = merged[["group", "squares"]].groupby("group").sum().rename(columns={"squares": "squares_sum"})
sqaures_sum["variance"] = sqaures_sum["squares_sum"] ** 0.5
```
